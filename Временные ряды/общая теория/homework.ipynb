```python
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import adfuller
import statsmodels.api as sm
import statsmodels.tsa.api as smt
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt
from scipy.stats import boxcox
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
%matplotlib inline

def test_stationarity(timeseries):
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    for key, value in dftest[4].items():
        dfoutput['Critical Value (%s)' % key] = value
    print(dfoutput)

def tsplot(y, lags=None, figsize=(14, 8), style='bmh', title=''):
    test_stationarity(y)
    if not isinstance(y, pd.Series):
        y = pd.Series(y)
    with plt.style.context(style):
        plt.figure(figsize=figsize)
        layout = (4, 1)
        ts_ax = plt.subplot2grid(layout, (0, 0), rowspan=2)
        acf_ax = plt.subplot2grid(layout, (2, 0))
        pacf_ax = plt.subplot2grid(layout, (3, 0))

        y.plot(ax=ts_ax, color='blue', label='Original')
        ts_ax.set_title(title)

        smt.graphics.plot_acf(y, lags=lags, ax=acf_ax, alpha=0.05)
        smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax, alpha=0.05)

        plt.tight_layout()
    return

#    
files = [
    "daily-total-female-births-in-cal.csv",
    "weekly-closings-of-the-dowjones-.csv",
    "mean-monthly-air-temperature-deg.csv",
    "international-airline-passengers.csv",
    "monthly-boston-armed-robberies-j.csv",
    "monthly-sales-of-company-x-jan-6.csv"
]

#   
for file in files:
    print(f"\n{'='*50}")
    print(f"Analyzing {file}")
    print(f"{'='*50}")
    
    #  
    try:
        if file == "weekly-closings-of-the-dowjones-.csv":
            data = pd.read_csv(file)
            series = data['Close']
        else:
            data = pd.read_csv(file)
            if 'Count' in data.columns:
                series = data['Count']
            elif 'Deg' in data.columns:
                series = data['Deg']
            else:
                series = data.iloc[:, 1]  #   ,   
    except Exception as e:
        print(f"Error loading {file}: {e}")
        continue
    
    # 1.  
    print("\nOriginal series:")
    tsplot(series, title=f'Original - {file}')
    plt.show()
    
    # 2.  -
    try:
        print("\nAfter Box-Cox transformation:")
        series_boxcox = boxcox(series, 0)
        tsplot(series_boxcox, title=f'Box-Cox transformed - {file}')
        plt.show()
        series = series_boxcox
    except Exception as e:
        print(f"Box-Cox transformation failed for {file}: {e}")
        continue
    
    # 3.  
    print("\nAfter first differencing:")
    series_diff = series[1:] - series[:-1]
    tsplot(series_diff, title=f'First difference - {file}')
    plt.show()
    
    # 4.   (  )
    if 'monthly' in file.lower() or 'airline' in file.lower() or 'temperature' in file.lower():
        print("\nAfter seasonal differencing (12 months):")
        try:
            series_seasonal_diff = series[12:] - series[:-12]
            tsplot(series_seasonal_diff, title=f'Seasonal difference (12) - {file}')
            plt.show()
        except Exception as e:
            print(f"Seasonal differencing failed for {file}: {e}")
    elif 'weekly' in file.lower():
        print("\nAfter seasonal differencing (52 weeks):")
        try:
            series_seasonal_diff = series[52:] - series[:-52]
            tsplot(series_seasonal_diff, title=f'Seasonal difference (52) - {file}')
            plt.show()
        except Exception as e:
            print(f"Seasonal differencing failed for {file}: {e}")
    
    # 5.   ( + )
    if 'monthly' in file.lower() or 'airline' in file.lower() or 'temperature' in file.lower():
        print("\nAfter first and seasonal differencing:")
        try:
            series_diff_seasonal = series_diff[12:] - series_diff[:-12]
            tsplot(series_diff_seasonal, title=f'First + Seasonal difference - {file}')
            plt.show()
        except Exception as e:
            print(f"Combined differencing failed for {file}: {e}")
    elif 'weekly' in file.lower():
        print("\nAfter first and seasonal differencing:")
        try:
            series_diff_seasonal = series_diff[52:] - series_diff[:-52]
            tsplot(series_diff_seasonal, title=f'First + Seasonal difference - {file}')
            plt.show()
        except Exception as e:
            print(f"Combined differencing failed for {file}: {e}")
```
